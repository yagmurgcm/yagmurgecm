## **Imports**

```python
from google.colab import drive
drive.mount('/content/drive', force_remount=True)

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from os.path import join
import warnings
warnings.simplefilter('ignore')

%matplotlib inline


path_prefix = "/content/drive/My Drive/"
data_path = "DSA - HW"
```

**Interest Rate- Google Trends**

```python

fname = "faiztrends.csv"

df = pd.read_csv(join(path_prefix, data_path, fname), skiprows=1)
df.columns = ['date', 'interest']
df['date'] = pd.to_datetime(df['date'])

#create date
df['month'] = df['date'].dt.to_period('M')

# Group by month and calculate mean interest
monthly_interest = df.groupby('month')['interest'].mean().reset_index()

# Line chart
plt.figure(figsize=(15, 6))
plt.plot(monthly_interest['month'].astype(str), monthly_interest['interest'], marker='o', linestyle='-', color='pink')
plt.title('Monthly Search Interest for "faiz" (Line Chart)')
plt.xlabel('Month')
plt.ylabel('Search (0–100)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Bar chart
plt.figure(figsize=(15, 6))
sns.barplot(x=monthly_interest['month'].astype(str), y=monthly_interest['interest'], color='pink')
plt.title('Monthly Search Interest for "faiz" (Bar Chart)')
plt.xlabel('Month')
plt.ylabel('Search(0–100)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

**TCMB Monthly Inflation Rates**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


file_path = '/content/drive/MyDrive/DSA - HW/inflationTCMB.csv'
df = pd.read_csv(file_path)

# Convert Date column to datetime
df['Date'] = pd.to_datetime(df['Date'])

# Create a 'month' column as Year-Month
df['Month'] = df['Date'].dt.to_period('M')

# Group by month to get monthly inflation averages (if needed)
monthly_inflation = df.groupby('Month')['Annual Change (%)'].mean().reset_index()
monthly_inflation['Month'] = monthly_inflation['Month'].astype(str)  # for plotting

# LINE CHART
plt.figure(figsize=(15, 6))
plt.plot(monthly_inflation['Month'], monthly_inflation['Annual Change (%)'], marker='o', linestyle='-', color='purple')
plt.title('Monthly Annual Inflation Change (Line Chart)')
plt.xlabel('Month')
plt.ylabel('Annual Change (%)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# BAR CHART
plt.figure(figsize=(15, 6))
sns.barplot(x=monthly_inflation['Month'], y=monthly_inflation['Annual Change (%)'], color='purple')
plt.title('Monthly Annual Inflation Change (Bar Chart)')
plt.xlabel('Month')
plt.ylabel('Annual Change (%)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

**Interest Rates TCMB**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file_path = '/content/drive/MyDrive/DSA - HW/faizTCMB.csv'
df = pd.read_csv(file_path)

#create date
df['Date'] = pd.to_datetime(df['Date'])
df['Month'] = df['Date'].dt.to_period('M')

# mean
monthly_rate = df.groupby('Month')['Interest'].mean().reset_index()
monthly_rate['Month'] = monthly_rate['Month'].astype(str)

#line chart
plt.figure(figsize=(15, 6))
plt.plot(monthly_rate['Month'], monthly_rate['Interest'], marker='o', linestyle='-', color='pink')
plt.title('Monthly TCMB Interest Rate (Line Chart)')
plt.xlabel('Month')
plt.ylabel('Interest Rate (%)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

#bar chart
plt.figure(figsize=(15, 6))
sns.barplot(x=monthly_rate['Month'], y=monthly_rate['Interest'], color='pink')
plt.title('Monthly TCMB Interest Rate (Bar Chart)')
plt.xlabel('Month')
plt.ylabel('Interest Rate (%)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
